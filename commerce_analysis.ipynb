## Convert Date Columns

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
```

```python
orders['order_month'] = orders['order_purchase_timestamp'].dt.to_period('M')
```

## Merge Datasets

```python
df = orders.merge(payments, on="order_id")
df = df.merge(customers, on="customer_id")
```

```python
df.head()
```

## Revenue Analysis

```python
df['payment_value'].sum()
```

## MONTHLY REVENUE 

```python
monthly_revenue = df.groupby('order_month')['payment_value'].sum()

monthly_revenue.plot(figsize=(10,5))
plt.title("Monthly Revenue Trend")
plt.show()
```

## Order Status Analysis

```python
orders['order_status'].value_counts()
```

```python
sns.countplot(x='order_status', data=orders)
plt.xticks(rotation=45)
plt.show()
```

## Customer Location Analysis

```python
customers['customer_state'].value_counts().head(10)
```

```python
sns.countplot(y='customer_state',
              data=customers,
              order=customers['customer_state'].value_counts().index[:10])
plt.show()
```

## Top Customers

```python
top_customers = df.groupby('customer_id')['payment_value'].sum().sort_values(ascending=False).head(10)

top_customers
```
